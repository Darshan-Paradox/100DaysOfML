## Adding and removing columns

```python
## Adding new column to existing data
df["bias"] = 1
df["age_norm"] = df["age"]/df.age.max()
```

```python
df.head()
```

```python
## removes the column bias
## axis = 0 represents rows and axis = 1 represents columns
## inplace = True makes changes to original DataFrame
df.drop("bias", axis = 1, inplace = True)
df.head()
```

## Dropping or Filling rows with empty, NaN or null element

```python
## Another way to remove NaN or null values
## set the how variable to any and subset to the chosen columns
(df.dropna(how="any", subset=["ip_address"])).head()
```

```python
## Filling missing values
(df.fillna(value = {"ip_address": "0.0.0.0"})).head()
```
