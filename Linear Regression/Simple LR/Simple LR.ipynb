```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv("./insurance.csv")
df.drop(columns=["sex", "bmi", "children", "region"], inplace=True)
```

```python
_df = df[df["smoker"] == "no"].drop(columns=["smoker"])
```

```python
_df = (_df - _df.min())/(_df.max() - _df.min())
_df.head()
```

```python
train = _df.sample(frac=0.8, random_state=1)
test = _df.drop(train.index)
```

```python
def grad_dscnt(X, Y, W, B, lr, iters):
    
    cost = []
    
    for i in range(iters):
        
        H = X.dot(W) + B
        
        n = len(Y)
        
        cost.append((1/(2*n))*np.sum(np.sum(Y - H)))
        
        W = W + lr*(1/n)*(Y - H).T.dot(X)
        B = B + lr*(1/n)*np.sum(Y - H)
    return cost
```

```python
X = np.array([train.age.values]).reshape(len(train.age.values), 1)
Y = np.array([train.charges.values]).reshape(len(train.charges.values), 1)
```

```python
W = np.random.rand(1, 1)
B = np.random.rand()
```

```python
cost = grad_dscnt(X, Y, W, B, 0.001, 50000)
```

```python
plt.plot(cost)
```

```python
H = X.dot(W) + B
```

```python
np.abs(H - Y).mean()
```

```python
_X = np.array([test.age.values]).reshape(len(test.age.values), 1)
_Y = np.array([test.charges.values]).reshape(len(test.charges.values), 1)

_H = _X.dot(W) + B
np.abs(_H - _Y).mean()
```
